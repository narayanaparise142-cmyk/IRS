```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
doc1="Machine learning is a branch of artificial intelligence that focuses on building applications that learn from data."
doc2="Artificial intelligence and machine learning are widely used to make smart applications based on data."
documents=[doc1,doc2]
vectorizer=TfidfVectorizer(stop_words='english')
tifdf_matrix=vectorizer.fit_transform(documents)
similarity_score=cosine_similarity(tifdf_matrix[0:1],tifdf_matrix[1:2])[0][0]
print("Cosine Similarity between Document1 and Document2:")
print(f"{similarity_score:.4f}")
```
