```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
text = "Text processing is an important task in natural language processing and data analysis."
words = nltk.word_tokenize(text)
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
processed_words = []
for word in words:
    word = word.lower()  # convert to lowercase
    if word.isalpha() and word not in stop_words:  # remove punctuation and stop words
        stemmed = stemmer.stem(word)  # stemming
        processed_words.append(stemmed)
print("Original Text:")
print(text)
print("\nProcessed Words (After Stopword Removal and Stemming):")
print(processed_words)
```
